### Average Final Grade (G3)

```python
average_grade = df['G3'].mean()
print("Average final grade:", round(average_grade, 2))
```

### Observation

The average final grade represents the overall academic performance of students in the dataset.

### Students' Scoring above 15

```python
high_score = df[df['G3'] > 15]

count = len(high_score)
percentage = (count/len(df))*100

print("Number of Students Scoring Above 15:", count)
print("Percentage of Students Scoring Above 15:", round(percentage, 2), "%")
```

```python
high_score
```

### Observation

Out of 395 students, **40 students (10.13%)** scored above **15** in the final grade (G3). This indicates that only a small percentage of students achieved excellent academic performance.

### Average Final Grade (G3) by Gender

```python
df.groupby('sex')['G3'].mean().round(2)
```

### Observation

The analysis shows that the average final grade (G3) of **male students is 10.91**, while **female students have an average grade of 9.97**. This indicates that male students performed slightly better than female students in this dataset. However, the difference is relatively small, suggesting that academic performance is fairly similar across both genders.

### Relationship between Study time and Final grade (G3)

```python
study_time = df.groupby('studytime')['G3'].mean().round(2).reset_index()
study_time
```

```python
study_time['studytime'] = study_time['studytime'].map({
    1: '<2 hrs',
    2: '2-5 hrs',
    3: '5-10 hrs',
    4: '>10 hrs'
})

fig = px.bar(
    study_time,
    x='studytime',
    y='G3',
    text='G3',
    title='Study Time vs Average Final Grade',
    labels={'studytime': 'Weekly Study Time', 'G3': 'Average Final Grade'}
)

fig.update_layout(width=800, height=500)
fig.show()
```

### Observation

The bar chart shows that students who studied **5–10 hours per week** achieved the highest average final grade (**11.40**). Students studying **more than 10 hours per week** also performed well, while those studying **less than 5 hours per week** obtained comparatively lower grades. Overall, the results suggest that consistent study habits are associated with better academic performance.

### Pass vs Fail distribution according Final Grade (G3)

```python
pass_fail = df['G3'].apply(lambda x: 'Pass' if x >= 10 else 'Fail')
result = pass_fail.value_counts()
result
```

```python
plt.figure(figsize=(6,4))

df['Result'] = pass_fail
ax = sns.countplot(df, x='Result', color='darkgrey')

total = len(df)

for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(
        percentage,
        (p.get_x() + p.get_width()/2, p.get_height()),
        ha='center',
        va='bottom'
    )

plt.title("Pass vs Fail Distribution")
plt.xlabel("Result")
plt.ylabel("Number of Students")

plt.show()
```

### Observation

The count plot shows that the majority of students successfully passed the course, accounting for **67.1%** of the total students, while **32.9%** did not meet the passing criteria. Although the overall pass rate is satisfactory, the proportion of failing students suggests that additional academic support could help improve student outcomes.

### Average Final Grade (G3) based on Previous Academic Failures

```python
df.groupby('failures')['G3'].mean().round(2)
```

```python
failure_grade = df.groupby('failures')['G3'].mean().round(2)

fig = px.bar(
      failure_grade,
      x=failure_grade.index,
      y=failure_grade.values,
      text=failure_grade.values,
      title='Previous Academic Failures vs Average Final Grade',
      labels={'x': 'Number of Previous Failures', 'y': 'Average Final Grade'}
  )

fig.update_traces(hovertemplate='Number of Previous Failures: %{x}<br>Average Final Grade: %{y}<extra></extra>')
fig.update_layout(width=800, height=500)
fig.show()
```

### Observation

The bar chart shows a clear decline in average final grades as the number of previous academic failures increases. Students with no previous failures achieved the highest average grade (**11.25**), while those with multiple failures scored significantly lower. This suggests that previous academic performance is an important factor influencing future academic success.

### Average Final Grade based on Parents' Education

```python
mother_edu = df.groupby('Medu')['G3'].mean().round(2)
father_edu = df.groupby('Fedu')['G3'].mean().round(2)

print("Average Final Grade by Mother's Education")
print(mother_edu)

print("\nAverage Final Grade by Father's Education")
print(father_edu)
```

```python
mother_count = df['Medu'].value_counts().sort_index()
father_count = df['Fedu'].value_counts().sort_index()

print(mother_count)
print(father_count)
```

```python
fig = go.Figure()

labels = ['None', 'Primary', '5th-9th', 'Secondary', 'Higher']

fig.add_trace(go.Scatter(
    x=labels,
    y=mother_edu.values,
    mode='lines+markers',
    name="Mother's Education",
    customdata=mother_count.values,
    hovertemplate=
        "<b>%{x}</b><br>" +
        "Average Grade: %{y:.2f}<br>" +
        "Students: %{customdata}<extra></extra>"
))

fig.add_trace(go.Scatter(
    x=labels,
    y=father_edu.values,
    mode='lines+markers',
    name="Father's Education",
    customdata=father_count.values,
    hovertemplate=
        "<b>%{x}</b><br>" +
        "Average Grade: %{y:.2f}<br>" +
        "Students: %{customdata}<extra></extra>"
))

fig.update_layout(
    title="Parents' Education vs Average Final Grade",
    xaxis_title="Parent's Education Level",
    yaxis_title="Average Final Grade (G3)"
)

fig.show()
```

### Observation

The line chart shows that students whose parents completed higher education generally achieved better average final grades than those whose parents had primary or secondary education. Although the **'None'** category has the highest average grade, it likely represents a very small number of students and should be interpreted with caution. Overall, the results suggest a positive association between higher parental education and students' academic performance.

### Average Final Grade (G3) based on Internet Access

```python
internet_grade = df.groupby('internet')['G3'].mean().round(2)
internet_grade
```

```python
plt.figure(figsize=(8, 5))
bars = plt.bar(internet_grade.index, internet_grade.values)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():.2f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title("Internet Access vs Average Final Grade")
plt.xlabel("Internet Access")
plt.ylabel("Average Final Grade (G3)")

plt.show()
```

### Observation

The bar chart shows that students with internet access at home achieved higher average final grades than those without internet access. This suggests that internet access may support learning by providing additional educational resources and study opportunities. However, academic performance depends on multiple factors and internet access should be viewed as a supportive resource rather than the sole factor influencing success.

### Correlation Heatmap

```python
plt.figure(figsize=(12,7))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    fmt='.2f',
    cmap='Pastel1'
)

plt.title('Correlation Heatmap')
plt.show()
```

### Observation

The heatmap indicates that **G1** and **G2** have the strongest positive correlation with the final grade (**G3**), making previous academic performance the best predictor of final results. In contrast, **previous academic failures** show a negative correlation with G3, while **study time** and **absences** have only a weak relationship with academic performance.

## **Key Insights**

- The average final grade of the students is **10.42**, indicating that most students achieved a moderate level of academic performance.

- Students who studied **5–10 hours per week** achieved the highest average final grades, suggesting that consistent study habits contribute to better academic outcomes.

- Previous academic failures showed a clear negative impact on final grades, making them one of the strongest indicators of future academic performance.

- Students whose parents had higher education levels generally performed better, highlighting the positive influence of family educational background.

- Students with internet access achieved slightly higher average grades, suggesting that access to educational resources may support learning.

- The correlation analysis showed that **G1** and **G2** are the strongest predictors of the final grade (**G3**), while study time and absences have a comparatively weaker relationship with academic performance.

## **Recommendations**

- Encourage students to maintain consistent study habits, as regular study time is associated with improved academic performance.

- Identify students with previous academic failures at an early stage and provide timely academic guidance and support.

- Improve access to digital learning resources to promote independent learning and better academic outcomes.

- Encourage greater parental involvement in students' education, as family educational background appears to positively influence academic performance.

- Monitor students' early academic performance (G1 and G2) to identify those who may require additional support before the final examination.
